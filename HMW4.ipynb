```python
# Python script to plot what I need

import math
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

x = np.linspace(0,1,1000)
v = 0
a = [1,1,1,1,1,1,1,1]
# a = [i * 5 for i in a]
print(a)
for j in range(1,9):
    v = v + a[j-1]*np.sin(np.pi*j*x)
    
plt.plot(x,v)
plt.xlabel('Position(m)')
plt.ylabel('Velocity(m/s)')
plt.title('Plot of velocity vs position along string')
```

This is smaller towards the middle, because at x = 0.5m, each sine term becomes $sin(2*\pi*j)$, which will all be zero. Close to this point, all of the constituent sine terms will also be close to zero, which is why the function is close to zero around x = 0.5

```python

x = np.linspace(0,1,1000)
v = 0
a = [1,1,-1,-1,1,1,1,-1]
# a = [i * 5 for i in a]
print(a)
for j in range(1,9):
    v = v + a[j-1]*np.sin(np.pi*j*x)
    
plt.plot(x,v)
plt.xlabel('Position(m)')
plt.ylabel('Velocity(m/s)')
plt.title('Plot of velocity vs position along string (bump at x=0.4m)')
```

This is a plot of the signal when the following coefficients of a are used:
a = [1,1,-1,-1,1,1,1-1]

```python
# PLotting the final signal
y1 = 0
y2 = 0
y3 = 0
y4 = 0
y5 = 0
y6 = 0
x = np.linspace(0,1,1000)
for i in range(1,9):
    y1 = y1 + np.pi*np.sin(np.pi*i*0.0)*np.cos(np.pi*i*x)
    y2 = y2 + np.pi*np.sin(np.pi*i*0.4)*np.cos(np.pi*i*x)
    y3 = y3 + np.pi*np.sin(np.pi*i*0.8)*np.cos(np.pi*i*x)
    y4 = y4 + np.pi*np.sin(np.pi*i*1.2)*np.cos(np.pi*i*x)
    y5 = y5 + np.pi*np.sin(np.pi*i*1.6)*np.cos(np.pi*i*x)
    y6 = y6 + np.pi*np.sin(np.pi*i*2.0)*np.cos(np.pi*i*x)

fig, axs = plt.subplots(3, 2)
fig.suptitle('Images of the string at various times')
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('t = 0.0s')
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('t = 0.4s')
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('t = 0.8s')
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('t = 1.2')
axs[2, 0].plot(x, y5)
axs[2, 0].set_title('t = 1.6')
axs[2, 1].plot(x, y6)
axs[2, 1].set_title('t = 2.0')

for ax in axs.flat:
    ax.set(xlabel='x(m)', ylabel='y')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

We can see that the wave reflects at x = 1 and starts travelling the other way!

```python
# Trying a heatmap
def heatmap2d(arr: np.ndarray):
    #plt.imshow(arr, cmap='viridis')
    plt.imshow(arr, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title('Heatmap showing the position of y with respect to x and t')
    plt.show()
    
    
# We need to create the array of the signal 
results = np.zeros(100)
x = np.linspace(0,1,100)
y=0
t = np.linspace(0,3,100)
for tx in t:
    y = 0
    for i in range(1,9):
        y = y + np.pi*np.sin(np.pi*i*tx)*np.cos(np.pi*i*x)
    
    results = np.vstack([results,y])
    
heatmap2d(results)
```

The heatmap shows rhythmic oscillations of y as time goes on (time is represented on the vertical axis, and x is represented on the horizontal axis. This is quite a nice visual of the wave reflecting as it hits the boundaries!
